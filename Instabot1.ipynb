## 1) $\text{Log in using Instagram username and password}$

```python
def login(username,password):
    try:
        
        x=wait.until(EC.presence_of_element_located((By.NAME,'username')))
        
        #log in page
        txtbox=driver.find_element(By.NAME,'username')
        txtbox.send_keys(username)

        txtbox=driver.find_element(By.NAME,'password')
        txtbox.send_keys(password)

        driver.find_element(By.CLASS_NAME,'_acan._aiit._acap._aijb._acas._aj1-').submit()
        
        #save your info: No
        pop_up=wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_ac8f')))
        pop_up.click()

        
        #Turn on Notification: No
        pop_up=wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_a9--._a9_1')))
        pop_up.click()
    except TimeoutException:
        print('Oops, Time out')
        
```

## 2) $\text{Type for “food” in search bar}$
$\text{and print all the names of the Instagram Handles that are displayed in list after typing “food”}$

```python
def display():
    try:
        # waiting and finding search bar
        wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_aacl._aacp._aacu._aacx._aada')))
        driver.find_element(By.CLASS_NAME,'_aacl._aacp._aacu._aacx._aada').click()
        
        # waiting and typing 'food' in search bar
        wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_aauy')))
        driver.find_element(By.CLASS_NAME,'_aauy').send_keys('food')

        #give time to load all handles
        
        wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_abm4')))
        time.sleep(5)
        #printing the handles
        
        for i in driver.find_elements(By.CLASS_NAME,'_abm4'):
            if len(i.text)==0 or i.text.split('\n')[0][0]=='#':
                continue
            
            print(i.text.split('\n')[0])
            
        #closing the search bar
        driver.find_element(By.CLASS_NAME,'_ab6-').click()
    except TimeoutException:
        print('Time limit exceeded')
```

## 3) $\text{Opening Profile So Delhi}$

```python
def open_profile(keyword):
    try:
        time.sleep(2)
        wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_aacl._aacp._aacu._aacx._aada')))
        driver.find_element(By.CLASS_NAME,'_aacl._aacp._aacu._aacx._aada').click()
        
        wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_aauy')))
        driver.find_element(By.CLASS_NAME,'_aauy').send_keys(keyword)
        
        time.sleep(3)
        wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_abm4')))
        
        driver.find_element(By.CLASS_NAME,'_abm4').click()    
        
    except TimeoutException:
        print('Time limit exceeded')
        
```

## 4) $\text{Follow/Unfollow given handle}$ 

```python
def follow():
    try:
        wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_ab8w ._ab94._ab97._ab9h._ab9k._ab9p._ab9x._abcm')))
        txt=driver.find_element(By.CLASS_NAME,'_ab8w ._ab94._ab97._ab9h._ab9k._ab9p._ab9x._abcm').text
        
        if txt=='Follow':
            driver.find_element(By.CLASS_NAME,'_ab8w ._ab94._ab97._ab9h._ab9k._ab9p._ab9x._abcm').click()
        else:
            print('You are already following')
        
    except TimeoutException:
        print('Time Limit exceeded')
```

```python
def unfollow():
    try:
        wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_acan._aiit._acap._aijb._acat._aj1-')))
        bttn=driver.find_element(By.CLASS_NAME,'_acan._aiit._acap._aijb._acat._aj1-').click()
        txt=driver.find_element(By.CLASS_NAME,'_ab8w ._ab94._ab97._ab9h._ab9k._ab9p._ab9x._abcm').text
        if txt!='Follow':
            wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_abm4')))
            time.sleep(3)
            bttn=driver.find_elements(By.CLASS_NAME,'_abm4')[4]
            bttn.click()
        else:
            print('Already unfollowed')
       
    except TimeoutException:
        print('Time Limit exceeded')
```

## 6) $\text{Extract list of followers}$

 $\text{sodelhi followers}$

```python
def Extract_Followers1():
    try :
        # locating followers button and click on it
        xpath="//a[contains(@href,'/sodelhi/followers/')]"
        driver.find_element(By.XPATH,xpath).click()
        
        wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_aano')))
        ele=driver.find_element(By.CLASS_NAME,'_aano')
        for i in range(20):
            time.sleep(2)
            driver.execute_script("arguments[0].scrollTop=arguments[0].scrollHeight",ele)

        
        wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_ab8w._ab94._ab97._ab9f._ab9k._ab9p._ab9-._aba8._abcm')))
        time.sleep(2)
        li=driver.find_elements(By.CLASS_NAME,'_ab8w._ab94._ab97._ab9f._ab9k._ab9p._ab9-._aba8._abcm')
        c=0
        followers=[]
        for i in li:
            
            followers.append(i.text.split()[0])
            c=c+1
            if c==500:
                break
        
        
        wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_ab8w._ab94._ab99._ab9f._ab9m._ab9p._ab9y._abcm')))
        driver.find_element(By.CLASS_NAME,'_ab8w._ab94._ab99._ab9f._ab9m._ab9p._ab9y._abcm').click()
        return followers
    except TimeoutException :
        print("Something Went Wrong!")
```
